```python
match_df = df.merge(matches[['id','venue']],left_on='match_id',right_on='id')
```

```python
match_df.drop(columns='id').sample(5)
```

```python
match_df['balls_left']=300-(6*(match_df['over']-1)+match_df['ball'])
```

```python
tf=match_df
```

```python
match_df=match_df.groupby(['match_id','inning']).sum()['total'].reset_index()
match_df=match_df[match_df['inning']==1]
match_df
```

```python
tf.drop(columns='match_id')
```

```python
df1=tf.merge(match_df[['match_id','total']],left_on='id',right_on='match_id')
```

```python
df1
```

```python
df1=df1[df1['inning']==1]
```

```python
df1['player_dismissed']=df1['player_dismissed'].fillna("0")
df1['player_dismissed']=df1['player_dismissed'].apply(lambda x:x if x == "0" else "1")
df1['player_dismissed']=df1['player_dismissed'].astype('int')
wickets=df1.groupby('match_id_x').cumsum()['player_dismissed']
```

```python
df1['wickets']=wickets
df1
```

```python
df2=df1[df1['wickets']>10]
df2.shape
```

```python
df1['Score']=df1.groupby('match_id_x').cumsum()['total_x']
df1
```

```python
groups = df1.groupby('id')

match_ids = df1['id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=60).sum()['total_x'].values.tolist())
```

```python
group = df1.groupby('id')

match_id = df1['id'].unique()
last_ten = []
for id in match_ids:
    last_ten.extend(groups.get_group(id).rolling(window=60).sum()['player_dismissed'].values.tolist())
```

```python
df1['last_ten']=last_five
df1['last_ten_wickkets']=last_ten
```

```python
df1
```

```python
df1['crr']=(df1['Score']*6)/(300-df1['balls_left'])
```
